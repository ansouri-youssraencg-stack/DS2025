```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

selected_categorical_features = [
    'Marital Status',
    'Gender',
    'Scholarship holder',
    'Debtor',
    'Course'
]

# Determine grid size dynamically for better flexibility
num_features = len(selected_categorical_features)
num_cols = 3 # Let's aim for 3 columns
num_rows = math.ceil(num_features / num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(num_cols * 7, num_rows * 6)) # Adjust figsize as needed
axes = axes.flatten() # Flatten the 2D array of axes for easy iteration

for i, feature in enumerate(selected_categorical_features):
    sns.countplot(data=X, x=feature, palette='viridis', hue=feature, legend=False, ax=axes[i])
    axes[i].set_title(f'Distribution of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Count')
    # Corrected line: Use set_xticklabels for rotation and alignment
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45, ha='right')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

print("Consolidated bar plots for selected categorical features have been generated.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

selected_categorical_features = [
    'Marital Status',
    'Gender',
    'Scholarship holder',
    'Debtor',
    'Course'
]

# Determine grid size dynamically for better flexibility
num_features = len(selected_categorical_features)
num_cols = 3 # Let's aim for 3 columns
num_rows = math.ceil(num_features / num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(num_cols * 7, num_rows * 6)) # Adjust figsize as needed
axes = axes.flatten() # Flatten the 2D array of axes for easy iteration

for i, feature in enumerate(selected_categorical_features):
    sns.countplot(data=X, x=feature, palette='viridis', hue=feature, legend=False, ax=axes[i])
    axes[i].set_title(f'Distribution of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Count')
    # Corrected line: Use plt.setp for rotation and alignment to avoid UserWarning
    plt.setp(axes[i].get_xticklabels(), rotation=45, ha='right')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

print("Consolidated bar plots for selected categorical features have been generated.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

selected_continuous_features = [
    'Admission grade',
    'Age at enrollment',
    'Unemployment rate',
    'Inflation rate',
    'GDP',
    'Curricular units 1st sem (grade)',
    'Curricular units 2nd sem (grade)'
]

# Determine grid size dynamically for better flexibility
num_features = len(selected_continuous_features)
num_cols = 3 # Aim for 3 columns
num_rows = math.ceil(num_features / num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(num_cols * 7, num_rows * 5)) # Adjust figsize as needed
axes = axes.flatten() # Flatten the 2D array of axes for easy iteration

for i, feature in enumerate(selected_continuous_features):
    sns.boxplot(data=df_combined, x='Target', y=feature, palette='viridis', hue='Target', legend=False, ax=axes[i])
    axes[i].set_title(f'{feature} vs. Target Outcome')
    axes[i].set_xlabel('Target Outcome')
    axes[i].set_ylabel(feature)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

print("Consolidated box plots for selected continuous features vs. 'Target' have been generated.")
```
